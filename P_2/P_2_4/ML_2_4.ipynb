```python
import numpy as np
import matplotlib.pyplot as plt

a = 1

hn = a / (100 * np.log(100))

def p_bar_n(x, a, hn):
    if x < 0:
        return 0
    elif x <= a:
        return (1 - np.exp(-x / hn)) / a
    else:
        return (np.exp(a / hn) - 1) * np.exp(-x / hn) / a

x_values = np.linspace(0, 0.05, 1000)  # x in the range 0 ≤ x ≤ 0.05
p_values = [p_bar_n(x, a, hn) for x in x_values]

plt.figure(figsize=(10, 5))
plt.plot(x_values, p_values, label=f'$\hat{{p}}_n(x)$ with $h_n = {hn:.5f}$')
plt.title('Estimated Probability Density Function $\hat{p}_n(x)$')
plt.xlabel('x')
plt.ylabel('$\hat{p}_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
