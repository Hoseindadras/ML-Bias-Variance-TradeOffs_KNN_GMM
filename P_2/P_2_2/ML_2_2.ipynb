```python
import numpy as np
import matplotlib.pyplot as plt

# Define the corrected Parzen window estimator function
def parzen_estimator_corrected(x, a, h_n):
    if x < 0:
        return 0
    elif 0 <= x <= a:
        return (1/a) * (1 - np.exp(-x / h_n))
    else:  # x >= a
        return (1/a) * (np.exp(a / h_n) - 1) * np.exp(-x / h_n)

# Define the range of x and the values for a and h_n
x_values = np.linspace(-0.1, 2, 400)  # x range extended for visualization beyond a = 1
a = 1
h_n_values = [1, 1/4, 1/16]

# Plot p̅_n(x) for different values of h_n
plt.figure(figsize=(12, 8))
for h_n in h_n_values:
    p_n = [parzen_estimator_corrected(x, a, h_n) for x in x_values]
    plt.plot(x_values, p_n, label=f'h_n = {h_n}')

plt.title(r'Parzen Window Estimator $\bar{p}_n(x)$ for Different $h_n$')
plt.xlabel('x')
plt.ylabel(r'$\bar{p}_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
